## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

# Decision logic here - you write this part!
if expense >= monthly_budget * 0.50:
  print("Major purchase")
# If expense > 50% of budget: "Major purchase - think carefully"
elif expense >= monthly_budget * 0.25:
  print("significant expense")
# If expense > 25% of budget: "Significant expense"
elif expense >= monthly_budget * 0.10:
  print("Moderate expense")
# If expense > 10% of budget: "Moderate expense"
else:
  print("Minor expense - within budget")
# Otherwise: "Minor expense - within budget"
````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

# Your task: Calculate percentage and give encouraging feedback
if current_savings >= goal_amount * 1.00:
  print("Goal achieved! Consider a new goal.")
# 100%+: "Goal achieved! Consider a new goal."
elif current_savings >= goal_amount * 0.75:
  print("Almost there! You're doing great.")
# 75-99%: "Almost there! You're doing great."
elif current_savings >= goal_amount * 0.50:
  print("Halfway there - Keep it up!")
# 50-74%: "Halfway there - keep it up!"
elif current_savings >= goal_amount * 0.25:
  print("Good progress, stay consistent")
# 25-49%: "Good progress, stay consistent"
else:
  print("Just getting started - you've got this!")
# Under 25%: "Just getting started - you've got this!"
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
# Smart Purchase Advisor (Advanced Version)

# Get user inputs
user_name = input("Enter your name: ")
purchase_amount = float(input("Enter the price of the item you want to buy: "))
available_money = float(input("Enter how much money you currently have: "))
emergency_fund = float(input("Enter your emergency savings amount: "))
category = input("Enter purchase category (necessity/luxury): ").lower()

print("\n--- Smart Purchase Advice ---")

# Check affordability
if purchase_amount > available_money:
    print("Sorry", user_name + ", you cannot afford this purchase right now.")
else:
    print("Good news", user_name + "! You can afford this purchase.")

    # Category-based advice
    if category == "necessity":
        print("This is a necessary purchase, so it may be reasonable to buy it.")
    elif category == "luxury":
        print("This is a luxury item. Consider whether you truly need it.")
    else:
        print("Unknown category. Try to decide if this item is essential or optional.")

    # Purchase size advice
    if purchase_amount < 50:
        print("This is a small purchase and usually low financial risk.")

    elif 50 <= purchase_amount <= 200:
        print("This is a medium purchase. Make sure it fits your monthly budget.")

    else:
        print("This is a large purchase. Careful financial planning is recommended.")

        # Payment plan options
        print("\n--- Payment Plan Options ---")
        for months in [3, 6, 12]:
            monthly_payment = purchase_amount / months
            print(months, "month plan: ", round(monthly_payment, 2), "per month")

    # Remaining money after purchase
    remaining_money = available_money - purchase_amount
    print("\nIf you buy this item, you will have", round(remaining_money, 2), "left.")

    # Low balance warning
    if remaining_money < 20:
        print("Warning: This purchase will leave you with very little spending money.")

    # Emergency fund protection rule
    if purchase_amount > (0.5 * emergency_fund):
        print("Warning: This purchase uses a large portion of your emergency fund.")
    else:
        print("Your emergency savings remain mostly protected.")

print("\nThank you for using the Smart Purchase Advisor!")
````

### Sample Decision Logic to Implement

- **Can't afford**: "Save up first - you need $X more"
- **Can afford but expensive**: "You can afford it, but it's Y% of your available money. Consider waiting."
- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"
- **Small purchase**: "This is a minor expense - go for it!"
