```python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# Expense entry
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Expense categorisation
food_keywords = ["food", "restaurant", "groceries", "snack"]
transport_keywords = ["bus", "taxi", "fuel", "transport", "uber"]
entertainment_keywords = ["movie", "game", "concert", "entertainment"]

description = expense_description.lower()

if any(word in description for word in food_keywords):
    category = "Food"
elif any(word in description for word in transport_keywords):
    category = "Transport"
elif any(word in description for word in entertainment_keywords):
    category = "Entertainment"
else:
    category = "Other"

print(f"\nExpense Category: {category}")

# Budget percentage calculation
percentage = (expense_amount / monthly_income) * 100

print(f"This expense represents {percentage:.2f}% of your monthly income.")

# Helpful feedback
if percentage < 5:
    print("✅ Good job! This is a small expense compared to your income.")
elif percentage < 20:
    print("⚠️ Moderate expense. Keep tracking your spending.")
else:
    print("❗ Warning: This is a large expense. Consider reviewing your budget.")

print("\nThank you for using Personal Finance Tracker!")
```
