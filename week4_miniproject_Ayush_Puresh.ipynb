## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
my_name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {my_name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts and descriptions as dictionaries
descriptions = []  # Store what each expense was for (categories)
total_spent = 0

print("=== Expenses Categories ===")
print("Enter expense categories (type 'finish' when done):")
while True:  # Infinite loop - we'll break out when user types "finish"

    category = input("Expense Category (or 'finish'): ").strip()

    # Check if user wants to stop
    if category.lower() == 'finish':
        break

   # check if input is empty
    if category.strip() == "":
        print("Category cannot be empty. Please try again.")
    else:
        descriptions.append(category)
        print(f"Category '{category}' added. Current total categories is: {len(descriptions)}")


# Remember: check for "finish", validate non-empty input, give feedback
print("\nCategories entered:")
for i,cat in enumerate (descriptions, 1):
    print(f"{i}. {cat}")

# Removed: budget_percentage = (expense_input / monthly_budget) * 100 -- caused NameError

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Use sentinel value pattern to collect multiple expenses
while True:
    expense_input = input("Expense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        expense = float(expense_input)
        if expense >= 0:
            # Ask for category and validate it
            expense_category_input = input(f"Enter category from {descriptions} for this expense: ")
            if expense_category_input in descriptions:
                expenses.append({'amount': expense, 'category': expense_category_input}) # Store as dictionary
                total_spent += expense
                budget_percentage = (total_spent / monthly_budget) * 100 # Calculate here after total_spent is updated
                print(f"  Added ${expense:.2f} to category {expense_category_input} | Running total: ${total_spent:.2f} | Budget spent {budget_percentage:.1f}%")
                if total_spent > monthly_budget:
                  print("  You are over budget!")

            else:
                print(f"  Invalid category. Please choose from {descriptions}.")
        else:
            print("  Expense must be positive!")
    except ValueError:
        print("  Please enter a valid number or 'done'")

# ===== FINAL SUMMARY =====
print("\n Expense entry complete!")
print(f"Total expenses: {len(expenses)} | Total spent: ${total_spent:.2f}")
print("\n--- Expense Breakdown ---")
if expenses: # Check if there are any expenses before iterating
    for i, exp in enumerate(expenses, 1):
        print(f"{i}. Category: {exp['category']} | Amount: ${exp['amount']:.2f}")
else:
    print("No expenses entered.")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Your task: Find highest and lowest expenses
    # Use the min/max pattern from workshops
    highest_expense = expenses[0]["amount"]
    lowest_expense = expenses[0]["amount"]

    for exp in expenses:
        if exp["amount"] > highest_expense:
            highest_expense = exp["amount"]
        if exp["amount"] < lowest_expense:
            lowest_expense = exp["amount"]

    print(f"\n Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Your min/max finding code here


    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Your task: Categorize expenses by size
    # Count how many expenses fall into each category:
    # Large (>20% of budget), Medium (5-20%), Small (<5%)

    large_count = 0    # > 20% of budget
    medium_count = 0   # 5-20% of budget
    small_count = 0    # < 5% of budget

    # Your categorization loop here
    for exp in expenses:
        percent = (exp["amount"] / monthly_budget) * 100
        if percent > 20:
            large_count += 1
        elif percent >= 5:
            medium_count += 1
        else:
            small_count += 1


    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # Your task: Provide smart feedback based on the analysis
    # Give different advice based on spending patterns
    print("\n Smart Feedback:")

    budget_used = (total_spent / monthly_budget) * 100

    if budget_used > 100:
        print(" You are over budget! Consider reducing large expenses.")
    elif large_count > 2:
        print("You have multiple large expenses. Try spreading costs more evenly.")
    elif medium_count > small_count:
        print("Medium expenses dominate — track them closely, they add up fast.")
    elif small_count > 0 and budget_used < 50:
        print("Good job keeping expenses small! You’re managing your budget well.")
    else:
        print("Your spending pattern looks balanced. Keep monitoring regularly.")

    # Your feedback logic here
    # Consider: budget percentage used, number of large expenses, etc.
    print("\n Expense Breakdown:")
    for i, exp in enumerate(expenses, 1):
        print(f"{i}. Category: {exp['category']} | Amount: ${exp['amount']:.2f}")

else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Goal name
    name = ""
    while name.strip() == "":
        name = input("Enter goal name: ").strip()
        if name == "":
            print("Goal name cannot be empty!")
    #   - Get target amount (with validation)
    target = 0
    while target <= 0:
        try:
            target = float(input("Enter target amount (must be > 0): $"))
            if target <= 0:
                print("Target must be positive!")
        except ValueError:
            print("Please enter a valid number!")
    #   - Get current amount (with validation)
    current = -1
    while current < 0 or current > target:
        try:
            current = float(input(f"Enter current saved amount (0-{target}): $"))
            if current < 0:
                print("Current amount cannot be negative!")
            elif current > target:
                print("Current amount cannot exceed target!")
        except ValueError:
            print("Please enter a valid number!")
    #   - Store in the three lists
    goal_names.append(name)
    goal_targets.append(target)
    goal_current.append(current)

print(f"\n Saved {len(goal_names)} goals!")


print("\n Savings Goals Summary:")
for i in range(len(goal_names)):
    progress = (goal_current[i] / goal_targets[i]) * 100
    remaining = goal_targets[i] - goal_current[i]
    print(f"{i+1}. {goal_names[i]} | Target: ${goal_targets[i]:.2f} | Current: ${goal_current[i]:.2f} | Progress: {progress:.1f}% | Remaining: ${remaining:.2f}")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0

    lowest_progress = 101  # start higher than 100%
    lowest_index = -1

    for i in range(len(goal_names)):
        # Progress calculation
        progress = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
        remaining = goal_targets[i] - goal_current[i]
        months_to_completion = remaining / 100 if remaining > 0 else 0  # assuming $100/month savings

        print(f"\n--- {goal_names[i]} ---")
        print(f"Target: ${goal_targets[i]:.2f}")
        print(f"Current: ${goal_current[i]:.2f}")
        print(f"Progress: {progress:.1f}%")
        print(f"Remaining: ${remaining:.2f}")
        print(f"Estimated months to completion (saving $100/month): {months_to_completion:.1f}")

        # Update totals
        total_target += goal_targets[i]
        total_current += goal_current[i]

        # Track lowest progress goal
        if progress < lowest_progress:
            lowest_progress = progress
            lowest_index = i

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Find which goal needs most attention
    if lowest_index >= 0:
        print("\n Goal Needing Most Attention:")
        print(f"'{goal_names[lowest_index]}' has the lowest progress at {lowest_progress:.1f}%.")
        print("Advice: Focus extra savings here to balance your progress across goals.")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum([e['amount'] for e in expenses]):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Display expense summary
        print("\n--- Expense Summary ---")
        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i, exp in enumerate(expenses, 1):
                print(f"{i}. ${exp['amount']:.2f} - {exp['category']}")
            print(f"\nTotal: ${sum([e['amount'] for e in expenses]):.2f}")

    elif choice == "2":
        # Display goals progress
        print("\n--- Goals Progress ---")
        if len(goal_names) == 0:
            print("No goals recorded yet.")
        else:
            for i in range(len(goal_names)):
                progress = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/${goal_targets[i]:.2f} ({progress:.1f}% complete)")

    elif choice == "3":
        # Add single new expense
       print("\n--- Add New Expense ---")
       try:
           expense_amount = float(input("Enter expense amount: $"))
           if expense_amount < 0:
             print("Expense must be positive!")
             continue

           # Ask for category and validate/add it
           category = input(f"Enter category from {descriptions} (or new one): ").strip()

           # Validate category input
           if category == "":
               print("Category cannot be empty! Please try again.")
               continue # Go back to the main menu loop

           # If category not in list, add it
           if category not in descriptions:
               descriptions.append(category)
               print(f"New category '{category}' added.")

           # Save expense with category
           expenses.append({"amount": expense_amount, "category": category})
           # Recalculate total_spent and budget_percentage for updated display
           current_total_spent = sum([e['amount'] for e in expenses])
           budget_percentage = (current_total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0
           print(f"Added ${expense_amount:.2f} to {category}. New total: ${current_total_spent:.2f} | Budget spent {budget_percentage:.1f}%")

       except ValueError:
           print("Please enter a valid number!")


    elif choice == "4":
         # Budget status check
        print("\n--- Budget Check ---")
        total_spent = sum([e['amount'] for e in expenses])
        budget_used = (total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0
        remaining = monthly_budget - total_spent
        daily_budget_left = remaining / 30 if remaining > 0 else 0

        print(f"Budget used: {budget_used:.1f}%")
        print(f"Remaining: ${remaining:.2f}")
        print(f"Daily budget left (approx): ${daily_budget_left:.2f}")

        if budget_used > 100:
            print(" You are over budget! Cut back immediately.")
        elif budget_used > 75:
            print("Caution: You’ve used most of your budget. Spend carefully.")
        elif budget_used > 50:
            print("You’re halfway through your budget. Keep monitoring.")
        else:
            print("Great job! You’re managing your budget well.")

    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {my_name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum([e['amount'] for e in expenses]) if expenses else 0:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
