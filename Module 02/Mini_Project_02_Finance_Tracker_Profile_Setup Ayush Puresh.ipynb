## Step 5 – Write the Code

````python
# Start program
print("Welcome to the Budget Tracker")

# Ask for user's name
name = input("Enter your name: ")

# Ask for monthly income
income = input("Enter your monthly income: ")

# Ask for preferred currency
currency = input("Enter your preferred currency: ")

# Ask for main savings goal
goal = input("Enter your main savings goal: ")

# Create and display summary message
print("\nHello", name + "!")
print("Your monthly income is", currency + income + ".")
print("Your main savings goal is", goal + ".")
print("Let's manage your budget wisely!")

# End program
````
