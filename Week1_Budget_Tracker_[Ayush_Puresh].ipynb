### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# AI-GENERATED BUDGET TRACKER CODE
# Ask for monthly income
# Ask for monthly income
income = float(input("Enter your monthly income:"))

# Collect up to 3 expenses
expenses = []
for i in range(1, 4):
    entry = input(f"Enter expense #{i} (or press Enter to skip): ")
    if entry.strip() == "":
        break
    expenses.append(float(entry))

# Calculate remaining budget
total_expenses = sum(expenses)
remaining = income - total_expenses

# Output results
print("\n--- Budget Summary ---")
print(f"Total income: ${income:.2f}")
print(f"Total expenses: ${total_expenses:.2f}")
print(f"Remaining budget: ${remaining:.2f}")

# Warning if budget is negative
if remaining < 0:
    print("⚠️ Warning: Your budget is negative! You're spending more than you make.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:

- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: [Expense categories (food, transport, entertainment)]"
enhancement_prompt = """
[ Enhancement Challenge
Choose ONE feature to add using AI assistance:

🏷️ Expense categories (food, transport, entertainment)]
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
# [# Ask the user for their name
name = input("What is your name? ")

# Ask for their monthly income
income = float(input("What is your monthly income? "))

# Expense categories
categories = ["Food", "Transport", "Entertainment"]
expenses = {}

# Ask user to enter expenses for each category
for category in categories:
    amount = input(f"Enter your {category} monthly expense (or press Enter to skip): ")
    if amount == "":
        expenses[category] = 0
    else:
        expenses[category] = float(amount)

# Calculate total expenses and remaining balance
total_expenses = sum(expenses.values())
remaining = income - total_expenses

# Print personalized welcome message and summary
print(f"\nWelcome, {name}!")
print("\n--- Budget Summary ---")
print(f"Monthly Income: {income}")

print("\nExpenses:")
for category, amount in expenses.items():
    print(f"{category}: {amount}")

print(f"\nTotal Expenses: {total_expenses}")
print(f"Remaining Balance: {remaining}")

# Warning if budget is negative
if remaining < 0:
    print("⚠ Warning: Your budget is negative!")
else:
    print("✅ Your budget is within limits.")
```
